```python
import matplotlib.pyplot as plt
```

```python
#Le dossier ou il y a les fonctions crées
import sys
sys.path.append('../Functions')
```

```python
#Importation des fonctions utiles
from data_load import load_csv
from sampling import oversampling, combinesampling
from evaluation import cla_repo, conf_mat
```

```python
#Chargement des données préparées
X_path='../Données/X_scaled.csv'
X=load_csv(X_path)
Y_path='../Données/Y.csv'
Y=load_csv(Y_path)
```

```python
#Appel de la fonction oversampling (elle fait le surechantillonnage et applique la LogisticRegression)
Y_predict, Y_test=oversampling(X , Y, test_size=0.3, max_iter=1500)
```

```python
#Rapport de classification pour le resultat de la fonction oversampling
cla_repo(Y_test, Y_predict)
```

```python
#La matrice de confusion pour le résultat de la fonction oversampling
conf_mat(Y_test, Y_predict)
plt.savefig('../Figures/Matrice_de_Confusion_oversampling.png')
```

```python
#Appel de la fonction combinesampling (elle fait l'echantillonnage equilibré et applique la LogisticRegression)
Y_pr_ba, Y_te_ba=combinesampling(X , Y, test_size=0.3, max_iter=1500)
```

```python
#Rapport de classification pour le resultat de la fonction combinesampling
cla_repo(Y_te_ba, Y_pr_ba)
```

```python
#La matrice de confusion pour le résultat de la fonction combinesampling
conf_mat(Y_te_ba, Y_pr_ba)
plt.savefig('../Figures/Matrice_de_Confusion_combinesampling.png')
```
